```python
import folium
from folium.plugins import FloatImage

m = folium.Map(
    location=[-23.545373, -46.474276],
    tiles="CartoDB dark_matter",
    zoom_start=12,
    zoom_control=False
    )



icon_url = 'corinthians-ate-decada-de-70-logo.png'
icon = folium.features.CustomIcon(icon_url,icon_size=(28, 30))  # Creating a custom Icon

html="""
<html>
<head></head>
<body>
<h3><b>Sport Club Corinthians Paulista</b></h3>
<p><i>“O Corinthians vai ser o time do povo e o povo é quem vai fazer o time.”</i></p>
<p>Miguel Battaglia</p>

</body
</html>
"""


folium.Marker(
    location=[-23.545373, -46.474276],
    icon=icon,
    popup=html,
).add_to(m)



m.save('index.html')
m
```
